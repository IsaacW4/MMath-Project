## Kerr black holes, ergospheres, event horizons, fast and slow rotation black holes, how geodesics behave near black holes

### Kerr black hole time-like with changing momentum in different directions with and against the spin

```python
import numpy as np 

from einsteinpy.geodesic import Geodesic, Timelike, Nulllike
from einsteinpy.plotting import GeodesicPlotter, StaticGeodesicPlotter, InteractiveGeodesicPlotter
```

```python
# Initial Conditions
position1 = [3, np.pi / 2, 0.]
position2 = [1.8, np.pi / 2, 0.]
momentum1 = [0., 0., -2.]
momentum2 = [0., 0., 2.]
a = 0.99 # Extremal Kerr Black Hole, as fast as it can possibly spin
#a1 = 0.5 # Half spinning black hole
#a2 = 0.1 # Another mid point
end_lambda = 15.
step_size = 0.0005
time_like = True
return_cartesian = True
julia = True # Using Julia
```

```python
geod1 = Geodesic(
    position=position1,
    momentum=momentum1,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)

geod2 = Geodesic(
    position=position1,
    momentum=momentum2,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)


geod3 = Geodesic(
    position=position2,
    momentum=momentum1,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)

geod4 = Geodesic(
    position=position2,
    momentum=momentum2,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)
```

```python
igpl = StaticGeodesicPlotter(
    bh_colors=("black", "white"), # Colors for BH surfaces, Event Horizon & Ergosphere
    draw_ergosphere=True # Ergosphere will be drawn
)
```

### Usual setup with Kerr and geodesic starting outside Ergosphere and geodesic momentum starting against spin of black hole

```python
igpl.plot(geod1, color="indigo") # figsize is in inches
igpl.save("geod31.png")
igpl.show()
```

### Setup with Kerr and geodesic starting outside Ergosphere and geodesic momentum starting WITH spin of black hole, geodesic not entering Ergosphere

```python
igpl.clear()
igpl.plot(geod2, color="indigo") # figsize is in inches
igpl.save("geod32.png")
igpl.show()
```

### Setup with Kerr and geodesic starting INSIDE Ergosphere and geodesic momentum starting against spin of black hole

```python
igpl.clear()
igpl.plot(geod3, color="indigo") # figsize is in inches
igpl.save("geod33.png")
igpl.show()
```
