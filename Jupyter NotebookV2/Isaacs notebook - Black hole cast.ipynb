## Shadow being cast around a thin emission disk of a Schwarzchild black hole

### Changing the mass of the Schwarzchuld black hole

```python
import astropy.units as u

from einsteinpy.rays import Shadow
from einsteinpy.plotting import ShadowPlotter
```

```python
mass1 = 0.1 * u.kg
mass2 = 0.5 * u.kg
mass3 = 1.0 * u.kg
mass4 = 1.5 * u.kg
mass5 = 5.0 * u.kg
mass6 = 3.0 * u.kg
fov = 30 * u.km 
# What field of view is the user expecting
```

```python
shadow = Shadow(mass=mass1, fov=fov, n_rays=1000)
```

```python
obj = ShadowPlotter(shadow=shadow, is_line_plot=True)
obj.plot()
obj.show()
```

```python
obj = ShadowPlotter(shadow=shadow, is_line_plot=False)
obj.plot()
obj.show()
```
