## Exploring Kerr Extremal Time-like Constant Radius Orbit, using Julia Backend and StaticGeodesicPlotter

```python
import numpy as np 

from einsteinpy.geodesic import Geodesic, Timelike, Nulllike
from einsteinpy.plotting import GeodesicPlotter, StaticGeodesicPlotter, InteractiveGeodesicPlotter
```

```python
# Initial Conditions
position = [4, np.pi / 3, 0.]
momentum = [0., 0.767851, 2.]
a = 0.99 # Extremal Kerr Black Hole
end_lambda = 300.
step_size = 1.
return_cartesian = True
julia = True # Using Julia
```

```python
geod = Timelike(
    position=position,
    momentum=momentum,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    return_cartesian=return_cartesian,
    julia=julia
)
```

```python
sgpl = StaticGeodesicPlotter(
    bh_colors=("black", "white"), # Colors for BH surfaces, Event Horizon & Ergosphere
    draw_ergosphere=True # Ergosphere will be drawn
)
```

```python
sgpl.plot(geod, color="tomato", figsize=(8, 8)) # figsize is in inches
sgpl.show()
```

```python
sgpl.plot2D(geod, coordinates=(1, 2), figsize=(6, 6), color="tomato") # X vs Y
sgpl.plot2D(geod, coordinates=(2, 3), figsize=(6, 6), color="tomato") # Y vs Z
```

```python
sgpl.parametric_plot(geod, figsize=(12, 7), colors=("red", "green", "blue"))
sgpl.show()
```

```python
# Using nbagg matplotlib Jupyter back-end for Interactive Matplotlib Plots - Necessary for animations
%matplotlib nbagg
sgpl.animate(geod, interval=10, color="tomato")
```

## Kerr black holes, ergospheres, event horizons, fast and slow rotation black holes, how geodesics behave near black holes

### Kerr black hole null-like with changing momentum in different directions with and against the spin

```python
import numpy as np 

from einsteinpy.geodesic import Geodesic, Timelike, Nulllike
from einsteinpy.plotting import GeodesicPlotter, StaticGeodesicPlotter, InteractiveGeodesicPlotter
```

```python
# Initial Conditions
position1 = [3, np.pi / 2, 0.]
position2 = [1.8, np.pi / 2, 0.]
momentum1 = [0., 0., -2.]
momentum2 = [0., 0., 2.]
a = 0.99 # Extremal Kerr Black Hole, as fast as it can possibly spin
#a1 = 0.5 # Half spinning black hole
#a2 = 0.1 # Another mid point
end_lambda = 15.
step_size = 0.0005
time_like = False
return_cartesian = True
julia = True # Using Julia
```

```python
geod1 = Geodesic(
    position=position1,
    momentum=momentum1,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)

geod2 = Geodesic(
    position=position1,
    momentum=momentum2,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)


geod3 = Geodesic(
    position=position2,
    momentum=momentum1,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)

geod4 = Geodesic(
    position=position2,
    momentum=momentum2,
    a=a,
    end_lambda=end_lambda,
    step_size=step_size,
    time_like=time_like, # Necessary to switch between Time-like and Null-like Geodesics, while using `Geodesic`
    return_cartesian=return_cartesian,
    julia=julia
)
```

```python
igpl = InteractiveGeodesicPlotter(
    bh_colors=("black", "white"), # Colors for BH surfaces, Event Horizon & Ergosphere
    draw_ergosphere=True # Ergosphere will be drawn
)
```

```python
igpl.plot(geod1, color="indigo") # figsize is in inches
igpl.show()
```

```python
igpl.clear()
igpl.plot(geod2, color="indigo") # figsize is in inches
igpl.show()
```

```python
igpl.clear()
igpl.plot(geod3, color="indigo") # figsize is in inches
igpl.show()
```

```python
igpl.clear()
igpl.plot(geod4, color="indigo") # figsize is in inches
igpl.show()
```
